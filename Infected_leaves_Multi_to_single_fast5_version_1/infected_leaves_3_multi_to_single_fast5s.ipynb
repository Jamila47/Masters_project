***converting multi fast5 in mapped fastq to single fast5***

Tasks we need to get: *Read ID of all mapped fastqs *extract them into batches *convert them into single fast5s

**To perform the 1st task your input = mapped bam file multifast5 of fastqs that mapped to the reference genome make sure to define the input directories before running the codes

```python
import os
```

```python
BAMIN_DIR = '../../analyses/mapping/infected_leaves/infected_leaves_3/'
FAST5IN_DIR = '../../data/genomic_data/infected_leaves/workspace_fast5/infected_leaves_3_fast5_out/'
OUT_DIR = '../../analyses/single_fast5s/infected_leaves/infectedleaves_3/'
```

***Define directories by giving them absolute path instead of typing the pathway over and over again

```python
BAMIN_DIR = os.path.abspath(BAMIN_DIR)
FAST5IN_DIR =os.path.abspath(FAST5IN_DIR)
OUT_DIR =os.path.abspath(OUT_DIR)
```

```python
###catch the bam based on naming convention
inbam_fn = [os.path.join(BAMIN_DIR, x) for x in os.listdir(BAMIN_DIR) if x.endswith ('sorted.bam')][0]
```

```python
mappedids_fn = inbam_fn.replace('.bam', '.mappedids.txt')
!samtools  view -F 4  {inbam_fn} | cut -f 1 | sort | uniq > {mappedids_fn}
```

***inorder to hold the batches of fast5s files generated by using the upper command, we need a temperory folder

```python
TMPOUT_DIR = os.path.join(OUT_DIR, 'tmp')
if not os.path.exists(TMPOUT_DIR):
    os.makedirs(TMPOUT_DIR)
```

***The files will be generated in a temperory folder, then to extract fast5s in batches

```python
###extract fast5s in bacthes
subset_ids = os.path.basename(mappedids_fn).replace('.txt', '')
!fast5_subset -i {FAST5IN_DIR} -s {TMPOUT_DIR} -f {subset_ids} -l {mappedids_fn} --recursive
```

```python
!multi_to_single_fast5 -i {TMPOUT_DIR} -s {OUT_DIR} -t 10 --recursive
!touch {OUT_DIR}/single_fast5.done
```
