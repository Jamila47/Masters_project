#now do other reps they might be single end reads
*the argument -i tells salmon where the index is and the argument -l tells salmon to it should automatically determine the library type of the sequnecing reads.
*the -1 and -2 tells the salmon where to find the left and right reads for this sample.

```python
fastq_fn_list = [os.path.join(FASTQ_DIR, x) for x in os.listdir(FASTQ_DIR)]
```

```python
for n in range(0, len(fastq_fn_list), 1):
    if "_" not in os.path.basename(fastq_fn_list[n]):
    #do single end
        print(fastq_fn_list[n])
        #print(fastq_fn_list[n + 1])
         #generate outfolder for each germinatedspores replicate
        QUANT_OUT = os.path.join(OUT_DIR, 'quant', \
                os.path.basename(fastq_fn_list[n].split('.')[0]))
        if not os.path.exists(QUANT_OUT):
            os.makedirs(QUANT_OUT)
        !{salmon_exc} quant -i {salmon_index} -l A -r {fastq_fn_list[n]} -p 8 --validateMappings -o {QUANT_OUT}
```
